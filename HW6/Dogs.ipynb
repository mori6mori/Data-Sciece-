## Q4 [12 Points]

Let's define a term called "gender-neutral". A dog name is gender-neutral if between 35% and 65% (including 35% and 65%) of the dogs using the name are of the same gender. 

* Report the **top 10 popular gender-neutral names** based on the frequency.

```python
nameFreq = pd.DataFrame (Dog3.groupby("AnimalName").size())

MaleFreq = pd.DataFrame (Dog3[Dog3.AnimalGender == 'M'].groupby("AnimalName").size())
FemaleFreq = pd.DataFrame (Dog3[Dog3.AnimalGender == 'F'].groupby("AnimalName").size())
#freq = pd.merge(nameFreq, MaleFreq, how="left", left_index=True, right_index=True)
freq =pd.concat([nameFreq, MaleFreq, FemaleFreq],axis=1, join="outer", ignore_index=False)
freq.columns = ["frequency","male_freq", "female_freq"]
display(freq)
```

```python

freq_neutral = freq.assign(percentage=freq.male_freq/freq.frequency)
#display(freq_neutral)

F4 = freq_neutral["percentage"]<=0.65
F5 = freq_neutral["percentage"]>=0.35

freq_neutral = freq_neutral.loc[F4 * F5]

freq_neutral = freq_neutral.sort_values("frequency", ascending = False)
display(freq_neutral)
```

## Q6. [8 Points]

An anlyst is interesting in whether the registration get busier during certain months 

* Use a line plot with marker="o" to show the relationship between registration month at registration and the **percentage** of records belonging to the licence month.


For this question, do not worry about the difference in the number of days of each month.


```python
 #pd.DataFrame(Dog["LicenseIssuedMonth"].value_counts()).reset_index
```

```python
month = pd.DataFrame(Dog.groupby("LicenseIssuedMonth").size()).reset_index()
month.columns = ['Month', 'LisenceIssued']
percentage = month.LisenceIssued/month.LisenceIssued.sum()


figure, ax=plt.subplots()
ax.plot(month.Month, percentage, marker = "o")
ax.set_xlabel("Month") # add x axis label
ax.set_ylabel("Percentage of record licenced") # add y axis label
ax.set_title("Month vs. Percentage of Record Licenced ") # add title
ax.set_xticks(np.arange(0, 12.5, 1))
plt.show() # show graph
```
